### **`*Basic numpy Operations*`**

---



```python
# 1. np.array()
# ‚û§ Purpose: Creates a NumPy array from a list or tuple.
# ‚û§ Example:
# arr = np.array([1, 2, 3])
# üîç Why Use It: It‚Äôs the foundation of NumPy ‚Äî much faster than Python lists.

import numpy as np
arr=np.array([1,2,3])
arr
```

```python
# 2. np.arange()
# ‚û§ Purpose: Creates a range of values with a step.
# ‚û§ Example:
# np.arange(1, 10, 2)
# üîç Why Use It: Useful for loops, indexing, or generating numeric sequences.


np.arange(0,10,2)
```

```python
# 3. np.linspace()
# ‚û§ Purpose: Generates evenly spaced values between a start and end.
# ‚û§ Example:
# np.linspace(0, 1, 5)
# üîç Why Use It: Best for graphs, simulations, and sampling.

np.linspace(0,1,5)
```

```python
# 4. np.zeros()
# ‚û§ Purpose: Creates an array of zeros.
# ‚û§ Example:
# np.zeros((3,3))
# üîç Why Use It: For initializing matrices before filling data.

np.zeros((3,3))
```

```python
# 5. np.ones()
# ‚û§ Purpose: Creates an array of ones.
# ‚û§ Example:
# np.ones(5)
# üîç Why Use It: Good for placeholders or constants.

np.ones(5)
```

```python
# 6. np.eye()
# ‚û§ Purpose: Generates an identity matrix.
# ‚û§ Example:
# np.eye(4)
# üîç Why Use It: Important in algebra, ML, and matrix calculations.

np.eye(4)
```

```python
# 7. np.reshape()
# ‚û§ Purpose: Changes array shape without changing the data.
# ‚û§ Example:
# arr.reshape(2, 3)
# üîç Why Use It: Useful in ML models (inputs reshape).
arr = np.array([1, 2, 3,5,8,10])
arr.reshape(3,2)
```

```python
# 8. np.sum()
# ‚û§ Purpose: Adds all elements.
# ‚û§ Example:
# np.sum(arr)
# üîç Why Use It: Basic but essential for calculations.

np.sum(arr)
```

```python
# 9. np.mean()
# ‚û§ Purpose: Calculates average value.
# ‚û§ Example:
# np.mean(arr)

np.mean(arr)
```

```python
# 10. np.std()
# ‚û§ Purpose: Computes standard deviation.
# ‚û§ Example:
# np.std(arr)
# üîç Why Use It: Measures data spread.

np.std(arr)
```

```python
# 11. np.min() / np.max()
# ‚û§ Purpose: Finds smallest or largest value.
# ‚û§ Example:
# np.max(arr)

np.min(arr)
```

```python
np.max(arr)
```

```python
# 12. np.random.rand()
# ‚û§ Purpose: Generates random numbers between 0 and 1.
# ‚û§ Example:
# np.random.rand(3)

np.random.rand(3,2)
```

```python
# 13. np.random.randint()
# ‚û§ Purpose: Random integers in a specified range.
# ‚û§ Example:
# np.random.randint(1, 100)
# üîç Why Use It: Used in games, ML random sampling.

np.random.randint(1,100)
```

```python
# 14. np.concatenate()
# ‚û§ Purpose: Combines arrays.
# ‚û§ Example:
# np.concatenate([a, b])
# üîç Why Use It: Merging datasets.
a=([4,5,6,5])
b=([8,6,7,5])
np.concatenate([a,b])
```

```python
# 15. np.unique()
# ‚û§ Purpose: Returns unique elements.
# ‚û§ Example:
# np.unique(arr)

arr=([1,1,5,5,4,4,9,10])
np.unique(arr)
```

```python
# 16. np.sort()
# ‚û§ Purpose: Sorts an array.
# ‚û§ Example:
# np.sort(arr)
# üîç Why Use It: Ordering data

arr=([4,8,8,2,9,1,20])
np.sort(arr)
```

```python
# 17. Array Indexing (arr[ ], arr[:, ])
# ‚û§ Purpose: Access specific rows, columns and slices.
# ‚û§ Example:
# arr[0]
# arr[:, 1]
# üîç Why Use It: Ultra-fast data access ‚Äî essential in ML.

arr=np.array([[4,8,2,9,1,20]])

print(arr[:,5])
```

```python
# 18. np.where()
# ‚û§ Purpose: Returns positions where condition is true.
# ‚û§ Example:
# np.where(arr > 10)
arr=np.array([4,8,2,9,1,20])
np.where(arr > 2)
```

```python
# 19. np.dot()
# ‚û§ Purpose: Matrix multiplication / dot product.
# ‚û§ Example:
# np.dot(a, b)

a=8
b=7
np.dot(a,b)
```

```python
# 20. np.transpose() / arr.T
# ‚û§ Purpose: Swaps rows and columns.
# ‚û§ Example:
# arr.T
# üîç Why Use It: Required in ML pipelines & data reshaping.

arr.T
```
