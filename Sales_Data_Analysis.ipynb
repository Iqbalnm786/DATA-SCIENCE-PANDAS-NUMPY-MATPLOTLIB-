```python
import pandas as pd
df=pd.read_csv("/content/sales - pandas 1.csv")
df
```

```python
#Clean missing values + duplicates
df.duplicated()
df.drop_duplicates()
```

```python
#Calculate total price
df['total']=df['Quantity']*df['Price']
df['total']
```

```python
#Find top customers & products
df.groupby('Customer Name')['Quantity'].sum()
```

```python
#Filter with query()
df.query("Location == 'Delhi'")
```

```python
#Filter with query()
df.query("Product == 'Laptop'")
```

```python
#Filter with query()
df.query("Price>=20000")
```

```python
print(df.to_string())
```

```python
#Filter with query()
df = df.rename(columns={'Customer Name': 'Customer_Name'})
df.query("Customer_Name == 'Anita'")
```

```python
# Apply custom functions
df['status']=df['total'].apply(lambda x: 'high' if x>=30000 else 'low' )
df
```
