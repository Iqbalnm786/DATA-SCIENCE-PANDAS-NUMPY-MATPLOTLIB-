# ***student_marks.csv***


```python
import pandas as pd
df=pd.read_csv("/content/student_marks-pandas 2.csv")
df.head(10)
```

```python
# Show number of students who scored grade A.
df.query('Grade=="A"')
```

```python
# Find average marks scored in each subject.

df.groupby('Subject')['Marks'].mean()
```

```python
#Which student scored the highest marks?

df.groupby(['Student'])['Marks'].max().head(1)
```

```python
# Add a new column "Pass/Fail"
# Marks ≥ 40 → Pass
# Else → Fail

df['Status']=df['Marks'].apply(lambda x:'Pass' if x >= 75 else 'Fail')
df
```

```python
# Sort marks in descending order.
df.sort_values(by='Marks',ascending=False)
```
